## Matplotlib example workflow

```python
#import matplotlib and get it ready
%matplotlib inline
import matplotlib.pyplot as plt

#prepare the data 
x=[1,2,3,4]
y=[11,22,33,44]

#setup the plot
fig, ax = plt.subplots(figsize=(10,10))

#plot the data
ax.plot(x,y)

#customize the plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

#save plot
fig.savefig("myplot.png")
```

## Two options for subplots

```python
# 1st option:creating subplots
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,5))
#plotting on each subplots
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(basketballers_range.keys(),basketballers_range.values());
ax4.hist(np.random.randn(1000));
```

```python
#creating option 2 for subpots
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
#plotting each axis subplots
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].barh(list(basketballers_range.keys()),list(basketballers_range.values()));
ax[1,1].hist(np.random.randn(1000));
```

```python
# explore a data frame
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"]= car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
#removing the last two zeros
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales
```

```python
car_sales["Price"].dtype
```

```python
#adding a sales date
car_sales["Sales date"]= pd.date_range("07/23/2022",periods=len(car_sales))
car_sales
```

```python
#adding up the total sales
car_sales["Total Sales"]=car_sales["Price"].cumsum()
car_sales
```

```python
#lets plot
car_sales.plot(x="Sales date", y="Total Sales");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist();
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot.scatter(x="age",
                    y="chol",
                    c="target");
```

```python
#OO METHOD
%matplotlib inline
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot.scatter(x="age",
                    y="chol",
                    c="target",
                    ax=ax);
ax.set_xlim([45,100]);
```

```python
#oo method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"]);

#customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

#add a legend to the plot
ax.legend(*scatter.legend_elements(),title="Target");

#plot the horizontal line of the mean of cholesterol
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
#subplots of age, chol, thalach

#CREATING THE SUBPLOTS
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True);
#ADDING DATA
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
#CUSTOMIZING THE PLOT
ax0.set(title="Heart Disease and Cholesterol levels",
        ylabel="Cholesterol");
#ADDING A LEGEND
ax0.legend(*scatter.legend_elements(),title="Target");

#GETTING A LINE TO SHOW THE MEAN
ax0.axhline(over_50["chol"].mean(),linestyle="--");

scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"]);
ax1.set(title="Heart disease and Max Heart rate",
       xlabel="Age",
       ylabel="Max HeartRate");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

#adding a heading to the overall figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

## This plot shows some information about the heart disease dataset

```python
#subplots of age, chol, thalach

#CREATING THE SUBPLOTS
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True);
#ADDING DATA
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");

#customizing or modyfing the range
ax0.set_xlim([50,80]);

#CUSTOMIZING THE PLOT
ax0.set(title="Heart Disease and Cholesterol levels",
        ylabel="Cholesterol");
#ADDING A LEGEND
ax0.legend(*scatter.legend_elements(),title="Target");

#GETTING A LINE TO SHOW THE MEAN
ax0.axhline(over_50["chol"].mean(),linestyle="--");

scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter");

#customizing or modyfing the range for ax1 
ax1.set_xlim([50,80])
ax1.set_ylim([60,200]);

ax1.set(title="Heart disease and Max Heart rate",
       xlabel="Age",
       ylabel="Max HeartRate");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

#adding a heading to the overall figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
